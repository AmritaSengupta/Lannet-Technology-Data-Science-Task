CREATING A USER DEFINED FUNCTION FOR THE DATE COLUMN

```
def DateFunction(data,day_first=True):
    col=data.columns   # Getting the column names of the data frame
    for i in col:      # creating a for loop to convert all the columns having date into a specified datetime format
        data[str(i)]=pd.to_datetime(data[str(i)],errors='coerce',dayfirst=day_first) # invalid dates will be converted to NaT
        
    for k in range(len(col)+1): # adding columns where the difference of the date columns will be stored
        if k<(len(col)-1):
            data[str(k+1)+'-'+str(k+2)]=data.iloc[:,k].subtract(data.iloc[:,k+1]) # subtracting two columns of the data frame
        else:
            data[str(len(col)-k)+'-'+str(k+1)]=data.iloc[:,k-(len(col)-1)]-data.iloc[:,k] # subtracting the first and last column of the dataframe
            break  # breaking the loop as soon as the condition is met
    print(data)
```

```
DateFunction(data=df1,day_first=True)
```

CREATING A USER DEFINED FUNCTION TO DROP THE VARIABLES FROM A DATAFRAME BASED ON THE PEARSON CORRELATION

```
def PearsonCorrelation(data):
    correlation=data.corr() # The default method is pearson for this correlation function 
    col=data.columns # Getting all the column names in the data frame
    for i in col:    # creating a for loop to filter out the columns whose pearson correlation is greater than 0.85 
        filter1= correlation[str(i)][correlation[str(i)] > 0.85]  # here we have not taken the absolute value of correlations because doing so will seggregate many more variables
        
        break  # breaking the loop as soon as we are meeting the condition
    dict1= dict(filter1)  # creating a dictionary of the filtered variables
    key=list(dict1.keys()) # getting the list of filtered column names in the form of keys
    print('Dropping columns:',key) 
    data=data.drop(columns=key) # dropping the filtered columns(those which are having correlation greater than 0.85)
    print(data.head()) # printing the new dataframe

```
